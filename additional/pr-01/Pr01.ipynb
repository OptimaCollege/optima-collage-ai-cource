```python
male = 0
female = 0

for indx in titanic.index:
    row = titanic.iloc[indx]
    if row["Gender"] == "male":
        male += 1
    elif row["Gender"] == "female":
        female += 1
    
print(f'Male: {male} \nFemale: {female}')
```
