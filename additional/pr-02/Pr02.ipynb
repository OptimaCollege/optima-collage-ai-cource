# Display data

Import libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
```

## Some simple charts

```python
x = [1, 2, 3, 4, 10]
y = [15, 5, 20, 3, 11]

plt.plot(y)
plt.show()
```

```python
plt.plot(x, y,color="red", marker="o")
plt.show()
```

```python
plt.plot(x, y,color="blue", marker="o")
y2 = [v * 2 for v in y]
plt.plot(x, y2,color="red", marker="^", linestyle="dashed")
plt.show()
```

```python
plt.scatter(x, y,color="red", marker="^")
plt.show()
```

**Read data**

```python
titanic = pd.read_csv("titanic.csv")
```

## Display

```python
l1 = ["Male", "Female"]
d1 = titanic["Gender"].value_counts().values
plt.bar(l1, d1)
plt.show()
```

```python
l2 = ["Children", "Adult", 'Unknown']

d2 = [0, 0, 0]
for d in titanic["Age"].values:
    if d < 18:
        d2[0] += 1
    elif d >= 18:
        d2[1] += 1
    else:
        d2[2] += 1

plt.bar(l2, d2, color='green')

# add text to bars
for i in range(len(d2)):
        plt.text(i, d2[i]//2, d2[i], ha = 'center')
        
plt.show()
```

```python
plt.pie(d2, labels=l2, autopct='%0.2f')       
plt.show()
```

```python
sns.barplot(x="Pclass", y="Survived", hue="Gender", data=titanic)
plt.show()
```

```python
sns.countplot(x="Gender", hue="Survived", data=titanic)
plt.show()
```

```python
sns.countplot(x="Embarked", hue="Survived", data=titanic)
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=titanic)
plt.show()
```
